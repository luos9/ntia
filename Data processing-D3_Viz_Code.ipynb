### For Section 1: Internet Usage by Map 

```python
#create list for columns and rows
states = entire_dataset.statename.unique()
internet_by_location = ['peinhome','peinwork','peinschl','peincafe','peintrav', 'peinlico', 'peinelho', 'peinothr']
```

```python
#calculate the internet usage by location for each state
def internet_by_loc(df,state,loc):
    state_df = df.loc[(df['statename']==state)&(df[loc]!=-1)]
    a=(state_df.pwsswgt).sum()
    state_df_loc = state_df[(state_df[loc] == 1)]
    b1 = state_df_loc.pwsswgt.sum()
    return b1/a
```

```python
#loop through all states in US
data_dic = {}
for i in states:
    loc_lst = []
    for j in internet_by_location:
        loc_lst.append(internet_by_loc(entire_dataset,i,j))
    data_dic[i] = loc_lst
```

```python
loc_df = pd.DataFrame.from_dict(data_dic, orient='index')
loc_df.columns = internet_by_location
loc_df
```

### For Section 2: Bar and Pie Chart

```python
def ageband(row):
    if row['prtage'] in range(7,23):
        return '07-22'
    elif row['prtage'] in range(23,39):
        return '23-38' 
    elif row['prtage'] in range(39,53):
        return '39-52'
    elif row['prtage'] in range(53,72):
        return '53-71' 
    elif row['prtage']> 72:
        return '72+'
#add a column for age band using prtage
entire_dataset['ageband'] = entire_dataset.apply(ageband, axis=1)
```

```python
def psex(row):
    if row['pesex'] == 1:
        return 'Male'
    elif row['pesex'] == 2:
        return 'Female' 
    elif row['prtage'] in range(39,53):
        return '39-52'
    elif row['prtage'] in range(53,72):
        return '53-71' 
    elif row['prtage']> 72:
        return '72+'
#add a column for age band using prtage
entire_dataset['ageband'] = entire_dataset.apply(ageband, axis=1)
```

```python
pie_df = entire_dataset.groupby(['statename','ageband'])
pie_df = pd.DataFrame(pie_df.size())
pie_df =pie_df.reset_index()
pie_df = pie_df.rename(index=str, columns={0: "age_count"})
```

```python
def count_age_loc(location,df,pie_df):
    filtered_df = df.loc[df[location] == 1]
    count_list = filtered_df.groupby(['statename','ageband']).count()[location].tolist()
    if len(count_list) == pie_df.shape[0]:
        pie_df.insert(loc=3, column=location+'_count', value=count_list)
```

```python
for i in internet_by_location:
    count_age_loc(i,entire_dataset,pie_df)
```
